```python
import pandas as pd
import sklearn as sk
```

```python
df = pd.read_csv('joensuu.csv', sep=' ')
```

```python
dp = sk.mixture.GaussianMixture(n_components=5, init_params='random')
```

```python
cls = dp.fit_predict(df.values)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df, x='x', y='y', c=cls)
```
